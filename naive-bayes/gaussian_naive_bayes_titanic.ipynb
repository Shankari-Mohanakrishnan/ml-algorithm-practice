# Gaussian Naive Bayes Classification using Titanic Dataset
## Objective

The goal of this notebook is to build a Gaussian Naive Bayes Classification model using the Titanic dataset and predict passenger survival.

## Dataset

The Titanic dataset contains passenger information from the Titanic ship disaster.

### Features Available:
- Passenger Class
- Gender
- Age
- Fare
- Family Information
- Cabin Information
- Embarked Port

### Target Variable:
- Survived
    - 0 → Did Not Survive
    - 1 → Survived

This is a binary classification problem.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

## Load Titanic Dataset

The Titanic dataset is loaded using pandas.

```python
df=pd.read_csv('titanic.csv')
df.head()
```

## Removing Unwanted Features

Some columns are removed because they are not useful for prediction or contain too many missing values.

Removed Columns:
- PassengerId
- Name
- SibSp
- Parch
- Ticket
- Cabin
- Embarked

```python
#removing unwanted features
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis=1, inplace=True)
df.head()
```

## Creating Independent and Dependent Variables

### Input Features
- Passenger Class
- Gender
- Age
- Fare

### Target Variable
- Survival Status

```python
#Dependent & Independent Variables
target=df.Survived
input=df.drop('Survived', axis=1)
```

## Handling Categorical Variables

Machine learning models cannot directly understand categorical text values like male and female.

`get_dummies()` is used to convert categorical columns into numerical format.


```python
#handling categorical columns 
dummies=pd.get_dummies(input.Sex)
dummies.head()
```

## Concatenating Dummy Variables

The encoded gender columns are combined with the original dataset.

```python
#dropping sex column
input.drop('Sex', axis=1,inplace=True)
input.head()
```

## Dropping Original Categorical Column

The original `Sex` column is removed after encoding because the model now uses the numerical dummy variables.

```python
input=pd.concat([input,dummies],axis=1)
input.head()
```

## Checking Missing Values

The dataset is checked for null values before model training.

The `Age` column contains missing values.

```python
#check for missing values 
input.isna().sum()
```

## Handling Missing Values

Missing values in the `Age` column are replaced using the mean age value.

This process is called Mean Imputation.

```python
input['Age']=input['Age'].fillna(input['Age'].mean())
```

```python
input.isna().sum()
```

## Train-Test Split

The dataset is divided into:

- Training Data → used to train the model
- Testing Data → used to evaluate model performance

A test size of 30% is used.

```python
#splitting data for training & prediction
X_train, X_test, y_train, y_test=train_test_split(input,target,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Gaussian Naive Bayes Classifier

Gaussian Naive Bayes is a supervised machine learning classification algorithm based on Bayes Theorem.

It assumes:
- features are independent
- numerical features follow Gaussian (Normal) distribution

GaussianNB works well for:
- small datasets
- fast classification tasks
- probabilistic predictions

```python
#creating a object of GaussianNB 
model=GaussianNB()
```

## Model Training

The Gaussian Naive Bayes model is trained using the training dataset.

```python
#fitting model
model.fit(X_train, y_train)
```

## Model Accuracy

The model achieved an accuracy score of approximately 81%.

This means the model correctly predicted survival outcomes for most passengers in the test dataset.

```python
#model scores - used to find accuracy 
model.score(X_test, y_test)
```

Internally does compare predicted and actual labels
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)

```python
X_test[:10]
```

## Prediction on Test Data

The trained model predicts whether passengers survived or not.

Predicted Output:
- 0 → Did Not Survive
- 1 → Survived

```python
y_test[:10]
```

```python
model.predict(X_test[:10])
```

## Probability Prediction

`predict_proba()` returns the probability of each class.

Example:
- Probability of not surviving
- Probability of surviving

The class with the highest probability becomes the final prediction.

```python
model.predict_proba(X_test[:10])
```

## Cross Validation

Cross-validation is used to evaluate model stability and generalization performance.

`cv=5` means:
- dataset is divided into 5 parts
- model is trained and validated 5 times

This helps reduce overfitting and gives a more reliable performance estimate.

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(), X_train, y_train, cv=5)
```

## Conclusion

The Gaussian Naive Bayes model successfully predicted passenger survival using the Titanic dataset.

### Key Observations

- GaussianNB is simple and computationally efficient
- Categorical variables must be converted into numerical format
- Missing values should be handled before training
- The model achieved approximately 81% accuracy
- Cross-validation helped evaluate model consistency
