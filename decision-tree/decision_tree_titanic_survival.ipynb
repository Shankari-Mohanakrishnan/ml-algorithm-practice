# Titanic Survival Prediction using Decision Tree Classifier

## Objective
The goal of this notebook is to predict passenger survival on the Titanic using a Decision Tree Classifier.

The model predicts whether a passenger survived based on:
- passenger class
- age
- gender
- fare amount

## Business Use Case
This type of classification workflow is similar to real-world predictive analytics problems such as:
- customer churn prediction
- loan default prediction
- fraud detection
- insurance claim prediction

```python
import pandas as pd
import numpy as np
```

## Dataset Overview

The Titanic dataset contains passenger information including:
- passenger class
- age
- gender
- fare
- survival status

Target Variable:
- Survived
  - 1 → survived
  - 0 → did not survive

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df.shape
```

## Feature Selection

Selected input features:
- Pclass
- Age
- Sex
- Fare

These variables are commonly associated with passenger survival probability.

```python
#splitting into dependent and independent variables
inputs=df[['Pclass','Age','Sex','Fare']]
inputs.head()
```

## Target Variable

The target variable is:
- Survived

This is a binary classification problem where the model predicts whether a passenger survived the disaster.

```python
target=df['Survived']
target
```

## Handling Categorical Variables

The 'Sex' column is categorical and cannot be directly used by machine learning algorithms.

One-hot encoding is applied using `pd.get_dummies()` to convert gender into numerical representation.

```python
sex_n=pd.get_dummies(inputs['Sex'])
sex_n
```

```python
inputs=pd.concat([inputs,sex_n], axis=1)
inputs.head()
```

## Avoiding Dummy Variable Trap

One dummy column ('male') is removed to avoid redundancy and multicollinearity.

The remaining 'female' column is sufficient to represent gender information.

```python
inputs.drop(['Sex','male'], axis=1, inplace=True)
inputs.head()
```

## Handling Missing Values

The 'Age' column contains missing values.

Missing ages are replaced using the mean age of the dataset to ensure the model can process all records.

```python
inputs['Age']=inputs['Age'].fillna(inputs['Age'].mean())
inputs.head(10)
```

## Train-Test Split

The dataset is divided into:
- Training dataset → used to train the model
- Testing dataset → used to evaluate model performance on unseen data

A test size of 20% is used.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Decision Tree Classification Model

Decision Tree Classifier is used for binary classification.

The algorithm works by:
1. Splitting data based on feature conditions
2. Creating decision rules
3. Predicting survival outcome based on learned patterns

Advantages:
- interpretable
- handles non-linear relationships
- easy to visualize

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

## Model Evaluation

The model is evaluated using accuracy score on the testing dataset.

Accuracy measures the percentage of correctly classified passenger outcomes.

```python
#accuracy
model.score(X_test,y_test)
```

## Conclusion

The Decision Tree model successfully predicted Titanic passenger survival using demographic and fare-related features.

Key observations:
- Gender and passenger class strongly influenced survival probability
- Missing values were handled using mean imputation
- Categorical variables were converted using one-hot encoding
- The model achieved approximately 78% accuracy on unseen data
